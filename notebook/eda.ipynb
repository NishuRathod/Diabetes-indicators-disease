```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set(style="whitegrid")
```

```python
data_path = Path("E:\Diabetes-indicators-disease\data")

files = list(data_path.glob("*.csv"))
print("Found CSV files:", files)

dfs = []
for f in files:
    df_temp = pd.read_csv(f)
    df_temp["source"] = f.name
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
df.info()
```

```python
print("Shape:", df.shape)
df.isna().sum()
```

```python
df.describe(include="all").T
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Diabetes_012')
plt.title("Target Distribution")
plt.show()
```

```python
num_cols = df.select_dtypes(include="number").columns

plt.figure(figsize=(12,10))
sns.heatmap(df[num_cols].corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[num_cols[:5]])  # First 5 numeric columns
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Diabetes_012", y="Age")
plt.title("Age distribution by Diabetes")
plt.show()
```

```python
num_cols = [c for c in df.columns if df[c].dtype != "object" and c != "Diabetes_012"]

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution: {col}")
    plt.show()
```

```python
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col)
    plt.title(f"Categorical Distribution: {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
df.to_csv("E:\Diabetes-indicators-disease\data\diabetes_merged_clean.csv", index=False)
print("Saved merged dataset.")
```
